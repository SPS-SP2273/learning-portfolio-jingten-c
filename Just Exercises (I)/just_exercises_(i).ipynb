## Exercise 4 :  A function for Fibonacci

```python
def fibonacci(n):
    filist=[0, 1]
    assert type(n) == int, "only integers larger than 0 allowed"
    assert n > 0, "n should be 1 or larger"
    if n == 1:
        return [0]
    for k in range (n - 2):
        newnum = filist[-1] + filist[-2]
        filist.append(newnum)
    return(filist)

fibonacci(1)
```

## Exercise 5 :  max_info() without NumPy

```python
def max_info(numbers):
    l = 0
    my_max = 0
    my_max_index = 0
    for i, test in enumerate(numbers):
        if test > my_max:
            my_max = test
            my_max_index = i
    return my_max, my_max_index
    
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
max_info(numbers)
```
